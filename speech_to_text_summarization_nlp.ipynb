```python
import re
import string
```

```python
def remove_numbers(text):
    result = re.sub(r'\d+', '', text)
    return result
```

```python
def remove_whitespace(text):
    return " ".join(text.split())
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

def remove_stopwords(text):
    stop_words = set(stopwords.words("english"))
    word_tokens = word_tokenize(text)
    filtered_word = [word for word in word_tokens if word not in stop_words]
    filtered_word = " ".join(filtered_word)
    return filtered_word
```

```python
def remove_punc(text):
    translator = str.maketrans('','',string.punctuation)
    return text.translate(translator)
```

```python
def text_lower(text):
    return text.lower()
```

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
ps = PorterStemmer()
def stemming(text):
    word_tokens = word_tokenize(text)
    filtered_word = [ps.stem(word) for word in word_tokens]
    filtered_word = " ".join(filtered_word)
    return filtered_word
```

```python
from nltk.stem import WordNetLemmatizer
lem = WordNetLemmatizer()
def lemetize(text):
    word_tokens = word_tokenize(text)
    filtered_word = [lem.lemmatize(word) for word in word_tokens]
    filtered_word = " ".join(filtered_word)
    return filtered_word
```

```python
# Removing special characters and digits
formatted_article_text = re.sub('[^a-zA-Z]', ' ', raw1 )
formatted_article_text = re.sub(r'\s+', ' ', formatted_article_text)
```

```python
import nltk
stopwords = nltk.corpus.stopwords.words('english')

word_frequencies = {}
for word in nltk.word_tokenize(formatted_article_text):
    if word not in stopwords:
        if word not in word_frequencies.keys():
            word_frequencies[word] = 1
        else:
            word_frequencies[word] += 1
```

```python
sentence_list = nltk.sent_tokenize(raw1)
```

```python
maximum_frequncy = max(word_frequencies.values())

for word in word_frequencies.keys():
    word_frequencies[word] = (word_frequencies[word]/maximum_frequncy)
```

```python
sentence_scores = {}
for sent in sentence_list:
    for word in nltk.word_tokenize(sent.lower()):
        if word in word_frequencies.keys():
            if len(sent.split(' ')) < 30:
                if sent not in sentence_scores.keys():
                    sentence_scores[sent] = word_frequencies[word]
                else:
                    sentence_scores[sent] += word_frequencies[word]
```

```python
import heapq
summary_sentences = heapq.nlargest(7, sentence_scores, key=sentence_scores.get)

summary = ' '.join(summary_sentences)
print(summary)

sum1=summary
```

```python
import nltk
import pandas as pd
from sentence_transformers import SentenceTransformer
from nltk.cluster import KMeansClusterer
import numpy as np
```

```python
model = SentenceTransformer('stsb-roberta-base')
```

```python
raw1 = raw2
```

```python
sentences=nltk.sent_tokenize(raw1)
# strip leading and trailing spaces
sentences = [sentence.strip() for sentence in sentences]
```

```python
data = pd.DataFrame(sentences)
data.columns=['sentence']
```

```python
def get_sentence_embeddings(sentence):
    embedding = model.encode([sentence])
    return embedding[0]
```

```python
data['embeddings']=data['sentence'].apply(get_sentence_embeddings)
```

```python
NUM_CLUSTERS=10
iterations=25
X = np.array(data['embeddings'].tolist())
kclusterer = KMeansClusterer(
        NUM_CLUSTERS, distance=nltk.cluster.util.cosine_distance,
        repeats=iterations,avoid_empty_clusters=True)
assigned_clusters = kclusterer.cluster(X, assign_clusters=True)
```

```python
data['cluster']=pd.Series(assigned_clusters, index=data.index)
data['centroid']=data['cluster'].apply(lambda x: kclusterer.means()[x])
```

```python
from scipy.spatial import distance_matrix
def distance_from_centroid(row):
    #type of emb and centroid is different, hence using tolist below
    return distance_matrix([row['embeddings']], [row['centroid'].tolist()])[0][0]
data['distance_from_centroid'] = data.apply(distance_from_centroid, axis=1)
```

```python
summary=' '.join(data.sort_values('distance_from_centroid',ascending = True).groupby('cluster').head(1).sort_index()['sentence'].tolist())
print(summary)
sum2=summary
```
